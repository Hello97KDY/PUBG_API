```python
url = 'https://api.pubg.com/shards/kakao/matches/eee04989-9f4f-40d4-a327-41f54330882b' 
# 첫번째 매치의 고유 id를 넣어준다
# 한 경기에 해당하는 모든 플레이어의 정보를 matches에 넣는다.
header = {
  "Authorization": "eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJqdGkiOiJlNDc4M2VkMC1jMDc5LTAxM2ItNzJhYS0yNjgwMDVjNzgxYzMiLCJpc3MiOiJnYW1lbG9ja2VyIiwiaWF0IjoxNjgxODY1MzM4LCJwdWIiOiJibHVlaG9sZSIsInRpdGxlIjoicHViZyIsImFwcCI6Ii05ZTZiMmJmMi02MmZkLTRjYmItYTE5Yi1lZjk2NTVhZmUzZDkifQ.kdD8DEO8nWROgcsbswLtTgOz7KgjYmd-IzEbv-De628",
  "Accept": "application/vnd.api+json"
}

r = requests.get(url, headers=header)
matches = r.json()
# print(matches)
r.json()
```

```python
# matches중에서 플레이어당 경기내용을 데이터프레임으로 만든다.

ilist = []
# name_list =[]
for i in range(len(matches['included'])):
      if matches['included'][i]['type'] == 'participant':
        ilist.append(matches['included'][i]['attributes']['stats'])
        matches['included'][i]['attributes']['stats']["id"] = matches['included'][i]['id'] # 개인별 id가 있기때문에 해당 user에 id도 넣어준다
#         name_list.append(matches['included'][i]['attributes']['stats']['name'])
matches_df = pd.DataFrame(ilist)
# df[['playerId']]
matches_df
```

```python
# roster안에 있는 팀원 id들을 모두 가져와 데이터 프레임으로 만든다.
# 이때 개인의 팀이 해당하는 rank를 각각 붙여준다.
id_list = []
for i in range(len(matches['included'])):
    if matches['included'][i]['type'] == 'roster':
        for k in range(len(matches['included'][i]['relationships']['participants']['data'])):
            matches['included'][i]['relationships']['participants']['data'][k]["rankx"] = matches['included'][i]['attributes']['stats']['rank']
            id_list.append(matches['included'][i]['relationships']['participants']['data'][k])
participant_id = pd.DataFrame(id_list)
participant_id
# d_list
```

```python
# 플레이어당 게임 데이터를 데이터프레임으로 만든 matches_df(개인별 id도 붙임) 와 개인별 id에 rank를 합친 participant_id 데이터프레임을
# 개인당 id를 기준으로 inner join 해준다
# 한경기에 해당하는 데이터프레임이 완성된다.

merdf = pd.merge(left = matches_df ,right = participant_id ,how = 'inner',on = 'id')
merdf
```
